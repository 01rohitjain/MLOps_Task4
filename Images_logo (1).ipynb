```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

# Create an image in OpenCV

```python
blank_img = np.zeros(shape=(500,500,3),dtype=np.int16)
blank_img=cv2.line(blank_img,pt1=(30,250),pt2=(250,470),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(30,250),pt2=(250,30),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(470,250),pt2=(250,470),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(470,250),pt2=(250,30),color=(105,105,105),thickness=5)

blank_img=cv2.line(blank_img,pt1=(40,240),pt2=(260,460),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(40,260),pt2=(260,40),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(460,240),pt2=(240,460),color=(105,105,105),thickness=5)
blank_img=cv2.line(blank_img,pt1=(460,260),pt2=(240,40),color=(105,105,105),thickness=5)

cv2.circle(img=blank_img, center=(250,250), radius=180, color=(105,105,105), thickness=3)
cv2.circle(img=blank_img, center=(250,250), radius=170, color=(105,105,105), thickness=3)

font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.putText(blank_img,text='Rohit',org=(140,230), fontFace=font,fontScale= 2.85,color=(105,105,105),thickness=7,lineType=cv2.LINE_AA)
cv2.putText(blank_img,text='Jain',org=(160,330), fontFace=font,fontScale= 2.5,color=(105,105,105),thickness=7,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
cv2.imwrite('blank_image.png', blank_img)
```
